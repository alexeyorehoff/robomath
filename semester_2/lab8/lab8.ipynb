```python
import numpy as np
from matplotlib import pyplot as plt

c = np.arange(0, 201, 10) 
measurements = np.array([101, 82, 91, 112, 99, 151, 96, 105])
log_odds = np.zeros_like(c, dtype=float)

def inverse_sensor_model(x, z):
    if x < z:
        return 0.3
    elif x <= z + 20:
        return 0.6
    else:
        return 0.5

for z in measurements:
    for idx, x in enumerate(c):
        p = inverse_sensor_model(x, z)
        log_odds[idx] += np.log(p / (1 - p))

m = 1 / (1 + np.exp(-log_odds))

plt.plot(c, m)
plt.xlabel('Distance from robot (cm)')
plt.ylabel('Occupancy Probability')
plt.grid(True)
plt.show()
```
